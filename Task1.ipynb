STEP 3: Data Preperation

Creating arrays for training and testing data



```python
x=pd.DataFrame(iris_DS,columns=["sepal_length","sepal_width","petal_length","petal_width"]).values
```

```python
x
```

```python
y=iris_DS.species.values.reshape(-1,1)
```

```python
y
```

Importing libraries

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

Train Model

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)  #coefficient of prediction
```

```python
model.intercept_
```

Making prediction


```python
y_pred=model.predict(x_test)
```

```python
print("Total mean squared Error: %.2f" % np.mean((y_pred-y_test)**2))
```
